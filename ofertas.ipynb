```python
from SIGAA import SIGAA
import pandas as pd

faltantes = pd.read_csv("data/Faltantes.csv", index_col=0)
faltantes
```

```python
materias_list = list([i.split("-")[0] for i in faltantes.index])
```

```python
equivalencias_nao_vazias = faltantes[
    faltantes["Equivalências"].notna() & (faltantes["Equivalências"] != "[]")
]["Equivalências"]
import itertools
import ast


# Remove valores nulos ou vazios
col = faltantes["Equivalências"].dropna()
col = col[col != "[]"].apply(ast.literal_eval)

lista_equivalencias = list(itertools.chain.from_iterable(col))
lista_equivalencias = list([i.split("-")[0] for i in lista_equivalencias])
```

```python
materias_list.extend(lista_equivalencias)
```

```python
import json

with open(r"data\2025_2.json", encoding="utf-8") as f:
    dados = json.load(f)

# Supondo que materias_list já esteja definido
# Exemplo: materias_list = ['ADMT0027 - ADMINISTRAÇÃO PARA ENGENHARIA', 'MAT001 - Cálculo I']

# Resultado armazenado aqui
ocorrencias = []

# Procura em todas as áreas (ex: COMPUTACAO, MECATRONICA, etc.)
for curso, componentes in dados.items():
    for componente in componentes:
        nome = componente.get("COMPONENTE CURRICULAR", "")
        for materia in materias_list:
            if materia.replace(" ", "").upper() in nome.replace(" ", "").upper():
                ocorrencias.append(
                    {
                        "Curso": curso,
                        "Componente": nome,
                        "Turma": componente.get("Turma", ""),
                        "Professor": componente.get("Professor", ""),
                        "SALA": componente.get("SALA", ""),
                        "Horários": componente.get("DIAS DA SEMANA e Horários", ""),
                    }
                )
                break  # Evita adicionar o mesmo componente várias vezes

# Exibe ou usa como quiser
df_ocorrencias = pd.DataFrame(ocorrencias)
df_ocorrencias.to_csv("data/possibilidades.csv", index=False, encoding="utf-8-sig")
```
