```python
import pandas as pd
df = pd.read_csv('data/2025.1/PDs/PDs.csv',header=0).dropna(how="all")
df
```

```python
df["Aula"].unique()
```

```python
materias_info = {
    'Controle da Qualidade': {'Sala': 'Simulações e Práticas Empresariais 1', 'Professor': 'Crisóstomo', 'Turma': 'P8'},
    'Calc Numérico': {'Sala': 'NT01', 'Professor': 'Sergio Floquet Sales', 'Turma': 'M4'},
    'Eng. Métodos': {'Sala': 'Sala 02', 'Professor': 'Angelo Leite', 'Turma': 'P4'},
    'Fis.Exp 3': {'Sala': 'Fis Exp 1', 'Professor': 'Raquel Aline Pessoa Oliveira', 'Turma': '4P'},
    'Fis 3': {'Sala': 'Sala 13', 'Professor': 'Militao Vieira Figueredo', 'Turma': 'M4'},
    'Metrologia': {'Sala': 'Sala 02', 'Professor': 'Gunther Costa', 'Turma': 'P4'},
    'Processos Industriais': {'Sala': 'LPQI', 'Professor': 'Vivianni Marques Leite dos Santos', 'Turma': 'PA'},
}
```

```python
# Adicionar informações de materias_info ao DataFrame
df['Sala'] = df['Aula'].map(lambda x: materias_info[x]['Sala'] if x in materias_info else None)
df['Professor'] = df['Aula'].map(lambda x: materias_info[x]['Professor'] if x in materias_info else None)
df['Turma'] = df['Aula'].map(lambda x: materias_info[x]['Turma'] if x in materias_info else None)
```

```python
import datetime

def get_weekday(date_str):
    date_obj = datetime.datetime.strptime(date_str, '%d/%m/%Y')
    return date_obj.strftime('%A')

df['Sala'] = df.apply(lambda row: 'Sala 05' if row['Aula'] == 'Metrologia' and get_weekday(row['Data']) == 'Thursday' else ('Sala 02' if row['Aula'] == 'Metrologia' and get_weekday(row['Data']) == 'Tuesday' else row['Sala']), axis=1)

df[df["Aula"] == "Metrologia"].head()
```

```python
# Salvar o DataFrame atualizado em um novo arquivo CSV
df.to_csv('data/2025.1/PDs/PDs_updated.csv', index=False)
df
```

```python
descricao_original = df['Descrição'].copy()

# Atualizar a coluna "Descrição" utilizando os valores originais
df['Descrição'] = df.apply(
    lambda row: f"{row['Professor']} \nTurma: {row['Turma']}\n{descricao_original.loc[row.name]}",
    axis=1
)

print(df['Descrição'][0])
```

```python
# Criar um novo DataFrame com as colunas necessárias para o Google Calendar
calendar_df = pd.DataFrame({
    'Subject': df['Aula'],
    'Start Date': df['Data'],
    'Start Time': df['Início'],
    'End Date': df['Data'],
    'End Time': df['Fim'],
    'All Day Event': 'False',
    'Description': df['Descrição'],
    'Location': df['Sala'],
    'Private': 'True'
})

# Salvar o DataFrame em um arquivo CSV
calendar_df.to_csv('data/2025.1/PDs/PDs_google_calendar.csv', index=False)
calendar_df.head()
```

```python
# Selecionar eventos cuja descrição contém "Atividade", "Avaliação" ou "Prova"
filtered_events = df[df['Descrição'].str.contains('Atividade|Avaliação|Prova', case=False, na=False)].copy()

# Converter a coluna "Data" para datetime
filtered_events['Data'] = pd.to_datetime(filtered_events['Data'])

# Ordenar os eventos utilizando o tipo datetime
filtered_events.sort_values(by='Data', inplace=True)

filtered_events
```
